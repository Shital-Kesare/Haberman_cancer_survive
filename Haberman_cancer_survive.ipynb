PDF ----> probability density function
CDF ----> cumulative distributive function


PDF looks at probability at one point whereas, CDF is the total probability of anything below it. 

CDF is the cumulative density function that is used for continuous types of variables.

```python
np.histogram(Haberman_df['age'],bins = 50)
```

```python
Haberman_df['age'].unique()
```

```python
count,bin_edges = np.histogram(Haberman_df['age'],bins = 10)

print('Count of occurance of Bins : ',count)
print('Bins created : ',bin_edges)

pdf = count/sum(count)
cdf = np.cumsum(pdf)

print('PDF  of  age',pdf)
print('CDF  of  age',cdf)
```

```python
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

```python
count,bin_edges = np.histogram(Haberman_df['year'],bins = 10)

print('Count of occurance of Bins : ',count)
print('Bins created : ',bin_edges)

pdf = count/sum(count)
cdf = np.cumsum(pdf)

print('PDF  of  year',pdf)
print('CDF  of  year',cdf)

plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

```python
count,bin_edges = np.histogram(Haberman_df['nodes'],bins = 10,density = True)

print('Count of occurance of Bins : ',count)
print('Bins created : ',bin_edges)

pdf = count/sum(count)
cdf = np.cumsum(pdf)

print('PDF  of  nodes',pdf)
print('CDF  of  nodes',cdf)

plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

Findings : from above figure if patient having less Nodes then high chances to survive. 

```python
patient_survived =Haberman_df.loc[Haberman_df['status'] == 1]
patient_not_survived = Haberman_df.loc[Haberman_df['status'] == 2]
```

```python
count,bin_edges = np.histogram(patient_survived['nodes'],bins = 10,density = True)

pdf = count/sum(count)
cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:],pdf,label = 'survival_pdf')
plt.plot(bin_edges[1:],cdf,label = 'survival_cdf')

count,bin_edges = np.histogram(patient_not_survived['nodes'],bins = 10,density = True)

pdf = count/sum(count)
cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:],pdf,label = 'not_survived_pdf')
plt.plot(bin_edges[1:],cdf,label = 'not_survived_cdf')

plt.title('CDF and PDF of survied and died class')
plt.xlabel('Nodes')
plt.xlabel('Probabilty')
plt.legend()
plt.show()
```

OBSERVATION :
    
1. From above survival_pdf that if patient has less than 10 nodes then patient has 82% chances of surviving
2. we can see that patient has both probability of surviving and dieing.
3. from not_survived_pdf we can see that around 58% patient died who as 10 or less than 10  nodes
